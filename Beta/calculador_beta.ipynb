```python
import pandas as pd
import yfinance as yf
import statsmodels.api as sm
from datetime import datetime, timedelta
```

```python
df_inside = pd.read_excel('carteirainside.xlsx')
ativos = []
```

```python
for ativo in df_inside['Ativo']:
    ativo = ativo + '.SA'
    ativos.append(ativo)
    
```

```python
ativos
```

```python
ativos.append('^BVSP')
```

```python
hoje = datetime.now()
cinco_anos_atras = hoje - timedelta(days=365)
```

```python
dados_cotacoes = yf.download(ativos, start=cinco_anos_atras, end=hoje)['Adj Close']
```

```python
retornos_diarios = dados_cotacoes.pct_change().fillna(0)
```

```python
x = retornos_diarios['^BVSP']
y_ativos = []
for ativo in retornos_diarios.columns:
    y = retornos_diarios[ativo]
    y_ativos.append(y)
```

```python
y_ativos
```

```python
df_final = []
for ativo in y_ativos:    
    X = sm.add_constant(x)
    model = sm.OLS(ativo, X)
    results = model.fit()
    df_final.append({'ativo': ativo.name, 'beta': results.params[1]})
```

```python
df_final
```

```python
df_final = pd.DataFrame.from_dict(df_final)
```

```python
# save df_final in xlsx file
df_final.to_excel('beta.xlsx')
```
