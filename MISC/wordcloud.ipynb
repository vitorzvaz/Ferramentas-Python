```python
from pypdf import PdfReader
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
reader = PdfReader("brit.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text()
```

```python
conectivos = set(STOPWORDS)
conectivos.update(["da", 'meu', 'em', 'e', 'você', 'de', 'ao', 'os', 'na', 'o', '4T23', 'para', 'milhões', 'resultado', 'que', 'nas', 'dos', 'n', 'p', '4T24', '3T24', 'foi', 'ano', 'Brisanet', 'resultados', 'trimestre', 'destaque', 'tabela', 'relatório', 'outro', 'anexo', 'outros', 'à', 'companhia', 'um', 'mais', 'release', 'não', 'pelo', 'por', 'ou', 'esse', 'essa', 'foram', 'uma'])
```

```python
nuvem_palavras = WordCloud(stopwords=conectivos, background_color='black', width=1920, height=1080).generate(text)
```

```python
fig, ax = plt.subplots(figsize= (20,20))

ax.imshow(nuvem_palavras)
ax.set_axis_off()
```
