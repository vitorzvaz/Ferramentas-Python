```python
import requests
import pandas as pd
```

```python
pular = 0

lista_dfs_moedas = []

while True:

    url_base = f"https://olinda.bcb.gov.br/olinda/servico/mecir_dinheiro_em_circulacao/versao/v1/odata/informacoes_diarias?$top=10000&$skip={pular}&$format=json"

    response = requests.get(url_base)

    data = response.json()

    df_moedas = pd.DataFrame(data['value'])

    lista_dfs_moedas.append(df_moedas)

    pular += 10000

    if len(data['value']) < 1:
        break

base_completa = pd.concat(lista_dfs_moedas, ignore_index=True)
```

```python
base_completa
```

```python
skip = 0
lista_dfs_moedas_comemorativas = []

while True:

    url_base = f"https://olinda.bcb.gov.br/olinda/servico/mecir_moedas_comemorativas/versao/v1/odata/informacoes_diarias?$top=10000&$skip={skip}&$format=json"

    response = requests.get(url_base)

    data = response.json()

    df_moedas_comemorativas = pd.DataFrame(data['value'])

    lista_dfs_moedas_comemorativas.append(df_moedas_comemorativas)

    skip += 10000

    if len(data['value']) < 1:
        break

base_completa_moedas_comemorativas = pd.concat(lista_dfs_moedas_comemorativas, ignore_index=True)
```

```python
base_completa_moedas_comemorativas
```

```python
# filter by 'Quantidade' < 100
base_completa_moedas_comemorativas[base_completa_moedas_comemorativas['Quantidade'] < 100].sort_values(by='Quantidade')
```
