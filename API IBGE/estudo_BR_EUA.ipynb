```python
import pandas as pd
import requests
import json
```

```python
paises = 'BR|US'
indicadores = '77827' # PIB
base_url = f'https://servicodados.ibge.gov.br/api/v1/paises/{paises}/indicadores/{indicadores}'
nome_pais = ['EUA', 'Brasil']

response = requests.get(base_url)
data = json.loads(response.text)

lista_pib = []

for i, serie in enumerate(nome_pais):

    lista_anos = []
    lista_valores = []

    for info in data[0]['series'][i]['serie']:

        valores = list(info.items())
        lista_anos.append(valores[0][0])
        lista_valores.append(valores[0][1])
    
    nome_coluna = nome_pais[i]
    df = pd.DataFrame(list(zip(lista_anos, lista_valores)), columns =['Ano', f"{nome_coluna}"]).dropna()

    lista_pib.append(df)
```

```python
df_pib = pd.merge(lista_pib[0], lista_pib[1], on='Ano')
df_pib = df_pib.set_index('Ano')
df_pib = df_pib.map(lambda x: float(x))
df_pib = df_pib / 1000000000 # Formato em Bilh√µes
```

```python
df_pib['Tamanho EUA'] = df_pib['EUA'] / df_pib['Brasil']
df_pib['Tamanho EUA'] = df_pib['Tamanho EUA'].map(lambda x: round(x, 2))
```
